```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
initial_dataframe=pd.read_excel('MDS- Case Study Data - Lifetime Value Prediction.xlsx')
```

```python
initial_dataframe['GROUP']=initial_dataframe['ACQ_TRIP_TYPE']+'/'+initial_dataframe['ACQ_BOOKING_PLATFORM']+'/'+initial_dataframe['ACQ_LENGTH_OF_STAY_CATEGORY']+'/'
+initial_dataframe['ACQ_BOOKING_LEADTIME']
```

```python
# initial_dataframe['LIFETIME_BEDVALUE']=initial_dataframe['LIFETIME_BEDVALUE'].replace(0,np.nan)
```

```python
# initial_dataframe['LIFETIME_BEDVALUE']= initial_dataframe['LIFETIME_BEDVALUE'].fillna(initial_dataframe.groupby('GROUP')['LIFETIME_BEDVALUE'].median())
```

```python
initial_dataframe=initial_dataframe[initial_dataframe['LIFETIME_BEDVALUE']!=0]
```

```python
initial_dataframe
```

```python
table = pd.pivot_table(initial_dataframe, values=['ACQ_BEDVALUE','CUSTOMER_ID'], index=['ACQ_BOOKEDDATE'],
                    columns=['GROUP'], aggfunc={'ACQ_BEDVALUE': np.mean,'CUSTOMER_ID':'count'})
table.columns = ["/".join(tup) for tup in table.columns.to_flat_index()]
table=round(table.reset_index(),2)
table
```

```python
groupbydf_date=initial_dataframe.groupby(['ACQ_BOOKEDDATE']).agg(ACQ_BEDVALUE_Mean=('ACQ_BEDVALUE', 'mean'), Customers_count =('CUSTOMER_ID', 'count'), LIFETIME_BEDVALUE_mean =('LIFETIME_BEDVALUE', 'mean'))
groupbydf_date=groupbydf_date.reset_index()
#groupbydf_date.set_index('ACQ_BOOKEDDATE_MONTH', inplace=True)
groupbydf_date
```

```python
sns.pairplot(groupbydf_date)
```

```python
merge1=groupbydf_date.merge(table, on=['ACQ_BOOKEDDATE'])
merge1=merge1.fillna(0)
merge1
```

```python
for i in [col for col in merge1 if col.startswith('ACQ_BEDVALUE/')]:
    merge1[i]=merge1[i]/merge1['ACQ_BEDVALUE_Mean']
```

```python
for i in [col for col in merge1 if col.startswith('CUSTOMER_ID/')]:
    merge1[i]=merge1[i]/merge1['Customers_count']
```

```python
merge1=round(merge1,4)
merge1
```

```python
merge1.to_csv('cohorts_7.csv')
```
