```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
initial_dataframe=pd.read_excel('MDS- Case Study Data - Lifetime Value Prediction.xlsx')
```

```python
initial_dataframe['GROUP']=initial_dataframe['ACQ_TRIP_TYPE']+'/'+initial_dataframe['ACQ_BOOKING_PLATFORM']+'/'+initial_dataframe['ACQ_LENGTH_OF_STAY_CATEGORY']+'/'
+initial_dataframe['ACQ_BOOKING_LEADTIME']
```

```python
initial_dataframe['LIFETIME_BEDVALUE']= initial_dataframe['LIFETIME_BEDVALUE'].fillna(initial_dataframe.groupby('GROUP')['LIFETIME_BEDVALUE'].median())
```

```python
initial_dataframe=initial_dataframe[initial_dataframe['LIFETIME_BEDVALUE']!=0]
```

```python
initial_dataframe[initial_dataframe['LIFETIME_BEDVALUE']==0]
```

```python
training_df=initial_dataframe
```

```python
training_df
```

```python
from sklearn import preprocessing
from sklearn.model_selection import train_test_split

le2 = preprocessing.LabelEncoder()
le2.fit(['Domestic', 'Short Haul', 'Long Haul'])
le2.classes_=(['Domestic', 'Short Haul', 'Long Haul'])
training_df['ACQ_TRIP_TYPE_encoded'] = le2.transform(training_df['ACQ_TRIP_TYPE'])

le3 = preprocessing.LabelEncoder()
training_df['ACQ_BOOKING_PLATFORM_encoded'] = le3.fit_transform(training_df['ACQ_BOOKING_PLATFORM'])

le4 = preprocessing.LabelEncoder()
le4.fit(['1-2 Nights', '3-4 Nights', '5-6 Nights','7+ Nights'])
le4.classes_=(['1-2 Nights', '3-4 Nights', '5-6 Nights','7+ Nights'])
training_df['ACQ_LENGTH_OF_STAY_CATEGORY_encoded'] = le4.transform(training_df['ACQ_LENGTH_OF_STAY_CATEGORY'])

le5 = preprocessing.LabelEncoder()
le5.fit(['Same day', '1-2 days', '3-6 days', '7-30 days', '31-60 days', '60+ days'])
le5.classes_=(['Same day', '1-2 days', '3-6 days', '7-30 days', '31-60 days', '60+ days'])
training_df['ACQ_BOOKING_LEADTIME_encoded'] = le5.transform(training_df['ACQ_BOOKING_LEADTIME'])

le6 = preprocessing.LabelEncoder()
training_df['GROUP_encoded'] = le6.fit_transform(training_df['GROUP'])

training_df.drop(['GROUP','ACQ_TRIP_TYPE','ACQ_LENGTH_OF_STAY_CATEGORY','ACQ_BOOKING_LEADTIME','ACQ_BOOKING_PLATFORM'],axis=1, inplace = True)

training_df

# training_df.drop(['ACQ_BOOKEDDATE'], axis=1, inplace = True)

# X = training_df.drop('LIFETIME_BEDVALUE',axis=1).values
# y = training_df['LIFETIME_BEDVALUE'].values

# X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```
