### At the end of the day, we are recieving a pivot dataframe, where our column represent combination of Timeband, BreakType (Block type) and Position Type in the block of commercials (may be extended if needed, in current case, limited due to computational complexity), rows = days, cell values are sum(GRP). Website visits are here as well. Dataset is limited by 14 days of TV ads campaign.

### From there we could run just simple linear regression, but what if we can complicate things a bit and apply adstock effect to each available collumn? Let's try this out

```python
exploration_df = merged_df [merged_df.Date_time_round.between(initial_dataframe1.Date_time_round.min(),initial_dataframe1.Date_time_round.max())]
exploration_df
```

```python
revenue_series = exploration_df.website_visits
decay_length = 30 #decay period
time = exploration_df.Date_time_round
```

```python
from scipy.optimize import minimize
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
```

```python
#
# Apply time lag (memory effect) to the campaign intensity series
# It is modeled as a convolution operation with exp decay filter (kernel)
#
def lag(x, alpha):
    w = np.array([ np.power(alpha, i) for i in range(decay_length) ])
    xx = np.vstack([ np.append(np.zeros(i), x[:len(x)-i]) for i in range(decay_length) ])
    y = np.dot(w/np.sum(w), xx)
    return y

#
# Apply the lag operation to each campaign/channel 
#
def adstock(x, alpha):
    return np.array([ lag(x[i], alpha[i]) for i in range(len(x)) ]).T

def evaluate_ols_loss(x, y, alpha):
    x_transformed = adstock(x, alpha)
    reg_model = LinearRegression().fit(x_transformed, y)
    return -reg_model.score(x_transformed, y)

#
# Optimize exp decay coeffecients jointly with the linear regression model
#
y = np.array(exploration_df['website_visits']).transpose()
x = np.array(exploration_df.drop(['Date_time_round','website_visits'],axis=1)).transpose()
solution = minimize(lambda alpha: evaluate_ols_loss(x, y, alpha), x0 = np.zeros(len(x)), tol=1e-6)
print(solution)
```

```python
plt.rcParams["figure.figsize"] = (20,8)
```

```python
#
# Compare model estimation with the actual revenue series
#
x_transformed = adstock(x, solution['x'])
reg_model = LinearRegression().fit(x_transformed, y)
website_visits_predicted = reg_model.predict(x_transformed)
plt.plot(time, y, time, website_visits_predicted)
```

```python
from sklearn.metrics import mean_squared_error

rms = mean_squared_error(y, website_visits_predicted, squared=False)
print ('RMSE: ', rms)
```

### Not the best fit, but at least some explainabilty

```python
df_coef = pd.DataFrame(exploration_df['Date_time_round'])
df_coef

for i in range(1,exploration_df.drop(['Date_time_round','website_visits'],axis=1).columns.shape[0]):
    coef_lin = reg_model.coef_[i] * x_transformed[:, i]
    df_coef[exploration_df.drop(['website_visits'],axis=1).columns[i]] = coef_lin
df_coef['intercept']=[reg_model.intercept_]*len(time)
```

```python
dict_df = dict(zip(df_coef.columns.sort_values(ascending=False),df_coef[df_coef.columns.sort_values(ascending=False)].drop('Date_time_round',axis=1).values.transpose()))
```

```python
# libraries
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# set seaborn style
sns.set_theme()
 
# Data
x= df_coef['Date_time_round']
y_to_plot= dict_df.values()
 
# Plot
plt.stackplot(x,y_to_plot, labels = dict_df.keys())
plt.legend(loc=(1.04,0))
plt.show()
```
