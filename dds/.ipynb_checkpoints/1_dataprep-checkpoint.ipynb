```python
import pandas as pd
import pandas_profiling
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from datetime import date
from datetime import datetime

%matplotlib inline
```

```python
initial_dataframe=pd.read_excel('exemplary_data.xlsx')
```

```python
initial_dataframe['godzina_time'] = pd.to_datetime(initial_dataframe['godzina'],format= '%H').dt.time
initial_dataframe['Data_date'] = pd.to_datetime(initial_dataframe['Data'] ).dt.date
initial_dataframe['Date_time_round'] =initial_dataframe.apply(lambda x: datetime.combine(x['Data_date'], x['godzina_time']), axis=1)
initial_dataframe['website_visits'] = initial_dataframe['Wizyty_all']
initial_dataframe['Date'] = initial_dataframe['Data']
#initial_dataframe['godzina'] = initial_dataframe['godzina'].astype(int)
#initial_dataframe.drop(['Wizyty_all','godzina_time'],inplace=True, axis=1)
initial_dataframe
```

```python
initial_dataframe1=pd.read_excel('exemplary_data.xlsx','spots')
```

```python
initial_dataframe1
```

```python
initial_dataframe1.loc[initial_dataframe1['Time'].str.startswith('24'), 'Time'] = initial_dataframe1['Time'].str.replace('24','00')
initial_dataframe1.loc[initial_dataframe1['Time'].str.startswith('25'), 'Time'] = initial_dataframe1['Time'].str.replace('25','01')
initial_dataframe1
```

```python
initial_dataframe1['Time_round']=initial_dataframe1['Time'].str.slice(0,2)

initial_dataframe1['Time_round']=pd.to_datetime(initial_dataframe1['Time_round'],format='%H').dt.time
initial_dataframe1['Date_time_round'] =initial_dataframe1.apply(lambda x: datetime.combine(x['Date'], x['Time_round']), axis=1)

initial_dataframe1
```

```python
print(initial_dataframe.Date_time_round.min(),initial_dataframe.Date_time_round.max())
print(initial_dataframe1.Date_time_round.min(),initial_dataframe1.Date_time_round.max())
```

```python
initial_dataframe1_cols_list = list(initial_dataframe1.drop(['Time','GRP','Title','Time_round','Date_time_round'],axis=1).columns)
initial_dataframe1_cols_list
```

```python
initial_dataframe_cols_list=list(initial_dataframe[['Date']].columns)
```

```python
gbdf1 = initial_dataframe1.groupby(['Date_time_round']).agg(GRP=('GRP', 'sum'),ads_count=('Title', 'count'))
gbdf1.reset_index(inplace=True)
gbdf1
```

```python
gbdf1.Date_time_round.value_counts()
```

```python
gbdf2 = initial_dataframe.groupby(initial_dataframe_cols_list).agg(website_visits=('website_visits', 'sum'))
gbdf2.reset_index(inplace=True)
gbdf2
```

```python
merged_df = pd.merge(gbdf1,gbdf2, on=['Date_time_round'],how ='right')
merged_df=merged_df.fillna(0.0)
merged_df
```

```python
f, ax = plt.subplots(figsize=(14, 8))

sns.lineplot(data=initial_dataframe,y='website_visits',x='Date_time_round')
ax.axvline(initial_dataframe1.Date_time_round.min(), color='r', zorder=-1)
ax.axvline(initial_dataframe1.Date_time_round.max(), color='r', zorder=-1)
ax.axvline(pd.to_datetime('2014-11-01 00:00:00'), color='g', zorder=-1) #prophet holiday
ax.axvline(pd.to_datetime('2014-11-11 00:00:00'), color='g', zorder=-1) #prophet holiday
ax.axvline(pd.to_datetime('2014-12-25 00:00:00'), color='g', zorder=-1) #chirstmas
```
